# Step 4: Feature Relationships
* Scatterplot
* Heatmap Correction
* Pairplot
* Groupby Comparisons

```python
df.plot(kind='scatter', x= 'Speed_Mph',y= 'Height_Ft', title= 'Coaster Speed vs Height')
plt.show()
```

Same things we do in seaborn its really look nice.

```python
sns.scatterplot(data= df,x='Speed_Mph',y='Height_Ft',hue='Year_Introduced',palette= 'coolwarm' )
plt.show()
```

```python
sns.pairplot(df, vars=['Year_Introduced','Speed_Mph','Height_Ft','Inversions_Clean','Gforce_Clean'], hue='Type_Main')
plt.show()
```

There we go. So we can see here on the right side the type of the material (red is wood, blue is other type and the purple is steel) and you could see that year introduced really early years there aren't much steel and then they start being introduced around the 1950s and then ever more popular. What else can we see here? A lot of interesting stuff we can see here just from this pair plot. So pair plots are really nice to use.

we want to do in comparing features against each other is to look at the correlation. So luckily in pandas it's very easy so we can run just on the subset of features that. we know are numeric and then we can let's go ahead and drop any null values and we can run on this just a corr function.

```python
df_corr =df[['Year_Introduced','Speed_Mph','Height_Ft','Inversions_Clean','Gforce_Clean']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr, annot=True )
```

What the correlation is Remember every value is going to be a perfect correlation with itself that's why we see ones across the diagonal. But otherwise this kind of lets us see interesting correlations and relationships in the data.
